```python
#define outliers
red_circle = dict(markerfacecolor= 'red', marker = 'o', markeredgecolor = 'white')
fig, axs = plt.subplots(1, len(X.columns), figsize=(20,10))
#enumerate: keep track of values
for i, ax in enumerate(axs.flat):
    ax.boxplot(X.iloc[:,i], flierprops = red_circle)
    ax.set_title(X.columns[i], fontsize  = 20, fontweight = 'bold', color = 'w')
    ax.tick_params(axis = 'y', labelsize=14, colors = 'white')
#space out
plt.tight_layout()
```

```python
cat_col = ['OverallQual', 'GarageCars', 'ExterQual', 'BsmtQual']
title = ['OverallQual', 'GarageCars', 'ExterQual', 'BsmtQual']

plt.figure(figsize=(7, 14))
for i, variable in enumerate(cat_col):
    plt.subplot(5,2,i+1)
    order = X[variable].value_counts(
        ascending=False).index   
    sns.set_palette('Set2')
    ax=sns.countplot(x=X[variable], data=X)
    sns.despine(top=True,right=True,left=True)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/len(X[variable]))
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        plt.annotate(percentage, (x, y),ha='center')
        plt.xticks(color = 'w')
        plt.yticks(color = 'w')
    plt.tight_layout()
    plt.title(title[i].upper(), color = 'w')
```
